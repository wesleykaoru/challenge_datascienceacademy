```python
# Importar as bibliotecas a serem usadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# Leitura dos datasets de treino e teste e criação do df de resposta
data = pd.read_csv("Taxas de Rendimento_ REGIOES_ 2007.csv", sep=";",encoding='latin-1')


# Concatenando os datasets
df = data[  ((data['Localização'] == "Total") & (data['Rede']== "Total") )]

         
features = [ 
'Região',
    
'Total Aprovação Fundamental',
'Total Reprovação Fundamental',
    
'Total  Aprovação Médio',
'Total  Reprovação Médio',
    
'Total Abandono Fundamental',
'Total Abandono - Médio'
]


ax = df['Total Aprovação Fundamental'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(16,8),color="teal");
ax = df['Total Reprovação Fundamental'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(16,8),color="coral");
ax = df['Total Abandono Fundamental'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(16,8),color="lightblue");

ax.set_title("% Aprovação / Reprovação / Abandono do Ensino Fundamental por REGIÃO", fontsize=18, y=1.04)
ax.set_xticklabels(df['Região'],rotation='horizontal', fontsize=15)
ax.legend(labels=['Aprovação','Reprovação','Abandono'], loc='upper left',fontsize=11)


aprovacao_fundamental=[]
for x in  df['Total Aprovação Fundamental'].str.replace(',', '.').astype(float):
    aprovacao_fundamental.append(x)

reprovacao_fundamental=[]
for x in  df['Total Reprovação Fundamental'].str.replace(',', '.').astype(float):
    reprovacao_fundamental.append(x)
    
abandono_fundamental=[]
for x in  df['Total Abandono Fundamental'].str.replace(',', '.').astype(float):
    abandono_fundamental.append(x)
    

# Atribuindo labels
for apr_fun,rep_fun,aba_fun,tam in zip(aprovacao_fundamental,reprovacao_fundamental,abandono_fundamental, range(0,5)):
    ax.text(x =tam+ -0.1, 
            y = apr_fun+0.8,
            s= apr_fun,
            fontsize=15)

    ax.text(x =tam+ -0.1, 
            y = rep_fun+0.8,
            s=rep_fun,
            fontsize=15)

    ax.text(x =tam+ -0.1, 
            y = aba_fun-1.21,
            s=aba_fun,
            fontsize=15)

# Salvando arquivo     
plt.savefig('C:\\Users\\WYWE\\Desktop\\Jupyter\\Data Science - Desafio\\REGIAO ensino_fundamental_.png', dpi=300, transparent=False)
```

```python
ax = df['Total  Aprovação Médio'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(18,10),color="teal", fontsize=12, width=0.75);
ax = df['Total  Reprovação Médio'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(18,10),color="coral", fontsize=12, width=0.6);
ax = df['Total Abandono - Médio'].str.replace(',', '.').astype(float).plot(kind='bar', figsize=(18,10),color="lightblue", fontsize=12, width=0.5);

aprovacao_medio=[]
for x_apr in  df['Total  Aprovação Médio'].str.replace(',', '.').astype(float):
    aprovacao_medio.append(x_apr)

reprovacao_medio=[]
for x_rep in  df['Total  Reprovação Médio'].str.replace(',', '.').astype(float):
    reprovacao_medio.append(x_rep)
    
abandono_medio=[]
for x_aba in  df['Total Abandono - Médio'].str.replace(',', '.').astype(float):
    abandono_medio.append(x_aba)
    

ax.set_title("% Aprovação / Reprovação / Abandono do Ensino Médio por REGIÃO", fontsize=18, y=1.04)
ax.set_ylabel("% Total", fontsize=18);
ax.set_xticklabels(df['Região'],rotation='horizontal', fontsize=15)
ax.legend(labels=['Aprovação','Reprovação','Abandono'], loc='upper left',fontsize=9)



# Atribuindo labels
for apr_med,rep_med,aba_med,tam in zip(aprovacao_medio,reprovacao_medio,abandono_medio, range(0,5)):
    ax.text(x =tam+ -0.0, 
            y = apr_med+0.8,
            s= apr_med,
            fontsize=15)

    ax.text(x =tam+ -0.0, 
            y = rep_med+2,
            s=rep_med,
            fontsize=15)

    ax.text(x =tam+ -0.0, 
            y = aba_med-3,
            s=aba_med,
            fontsize=15)

    
# Salvando arquivo    
plt.savefig('C:\\Users\\WYWE\\Desktop\\Jupyter\\Data Science - Desafio\\REGIAO ensino_medio_.png', dpi=300, transparent=False)
```
